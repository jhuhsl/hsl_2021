```python
#One function for all the tests
def basicStats(baseVAN, optiVAN, baseTAN, optiTAN):
    vertMis1 = [] #originally supposed to hold means for baseline, but it didn't work out how I expected (explained later)
    vertMis2 = [] #originally supposed to hold means for optivisor, but it didn't work out how I expected (explained later)
   
    #finding ocular misalignment
    for i in range(10):
        holderB = 0
        holderO = 0
        frame1 = baseVAN[i] # holds baseline trial from baseHold array
        frame2 = optiVAN[i] # holds optivisor trial from optiHold array
        for column in frame1[['YEnd']]:
            ang1 = 180/math.pi * np.arctan((frame1[column]/10)/207)
            ang2 = 180/math.pi * np.arctan((frame2[column]/10)/207)
            holderB += ang1
            holderO += ang2
        holderB /= 10
        holderO /= 10
        vertMis1.append(holderB) # idk why, but after this nested loop, both vertMis1 and vertMis2 hold every individual 
                                #test/10 instead of averages
        vertMis2.append(holderO)

    vertMisBaseMean = [] #holds baseline misalignment means for each trial
    vertMisOptiMean = [] #holds optivisor misalignment means for each trial
    vertMisBaseSD = [] #holds baseline misalignment standard deviation for each trial
    vertMisOptiSD = [] #holds optivisor misalignment standard deviation for each trial

    for l in range (10):
        vertMisBaseMean.append(vertMis1[l].sum()) #summing each dataframe with misalignment in the verMis1 and 2 arrays
        vertMisOptiMean.append(vertMis2[l].sum())
        vertMisBaseSD.append(vertMis1[l].std())
        vertMisOptiSD.append(vertMis2[l].std())

    #creating a dictionary of all the values I need for the table so I don't have to print stuff individually

    baseStats = {'Vertical Misalignment Means': vertMisBaseMean,
                 'Vertical Misalignment SD': vertMisBaseSD,
                 'Vertical Misalignment Final Mean': 0,
                 'Vertical Misalignment Final SD': 0,
                 'AngEnd Means': [baseTAN[0]['AngEnd'].mean(), 
                                  baseTAN[1]['AngEnd'].mean(), 
                                  baseTAN[2]['AngEnd'].mean(), 
                                  baseTAN[3]['AngEnd'].mean(), 
                                  baseTAN[4]['AngEnd'].mean(), 
                                  baseTAN[5]['AngEnd'].mean(), 
                                  baseTAN[6]['AngEnd'].mean(), 
                                  baseTAN[7]['AngEnd'].mean(), 
                                  baseTAN[8]['AngEnd'].mean(), 
                                  baseTAN[9]['AngEnd'].mean()], 
                 'AngEnd SD': [baseTAN[0]['AngEnd'].std(), 
                               baseTAN[1]['AngEnd'].std(), 
                               baseTAN[2]['AngEnd'].std(), 
                               baseTAN[3]['AngEnd'].std(), 
                               baseTAN[4]['AngEnd'].std(), 
                               baseTAN[5]['AngEnd'].std(), 
                               baseTAN[6]['AngEnd'].std(), 
                               baseTAN[7]['AngEnd'].std(), 
                               baseTAN[8]['AngEnd'].std(), 
                               baseTAN[9]['AngEnd'].std()],
                 'AngEnd Final Mean': 0, 
                 'AngEnd Final SD': 0}

    optiStats = {'Vertical Misalignment Means': vertMisOptiMean, 
                 'Vertical Misalignment SD': vertMisOptiSD,
                 'Vertical Misalignment Final Mean': 0,
                 'Vertical Misalignment Final SD': 0, 
                 'AngEnd Means': [optiTAN[0]['AngEnd'].mean(), 
                                  optiTAN[1]['AngEnd'].mean(), 
                                  optiTAN[2]['AngEnd'].mean(), 
                                  optiTAN[3]['AngEnd'].mean(), 
                                  optiTAN[4]['AngEnd'].mean(), 
                                  optiTAN[5]['AngEnd'].mean(), 
                                  optiTAN[6]['AngEnd'].mean(), 
                                  optiTAN[7]['AngEnd'].mean(), 
                                  optiTAN[8]['AngEnd'].mean(), 
                                  optiTAN[9]['AngEnd'].mean()],
                 'AngEnd SD': [optiTAN[0]['AngEnd'].std(), 
                               optiTAN[1]['AngEnd'].std(), 
                               optiTAN[2]['AngEnd'].std(), 
                               optiTAN[3]['AngEnd'].std(), 
                               optiTAN[4]['AngEnd'].std(), 
                               optiTAN[5]['AngEnd'].std(), 
                               optiTAN[6]['AngEnd'].std(), 
                               optiTAN[7]['AngEnd'].std(), 
                               optiTAN[8]['AngEnd'].std(), 
                               optiTAN[9]['AngEnd'].std()],
                 'AngEnd Final Mean': 0, 
                 'AngEnd Final SD': 0}


    #Finding mean values for vertical misalignment and AngEnd for both baseline and optivisor
    baseStats['Vertical Misalignment Final Mean'] = np.mean(baseStats['Vertical Misalignment Means'])
    baseStats['AngEnd Final Mean'] = np.mean(baseStats['AngEnd Means'])

    optiStats['Vertical Misalignment Final Mean'] = np.mean(optiStats['Vertical Misalignment Means'])
    optiStats['AngEnd Final Mean'] = np.mean(optiStats['AngEnd Means'])

    #Finding the final standard deviation based on the means of each trial
    baseStats['Vertical Misalignment Final SD'] = np.std(baseStats['Vertical Misalignment Means'])
    baseStats['AngEnd Final SD'] = np.std(baseStats['AngEnd Means'])

    optiStats['Vertical Misalignment Final SD'] = np.std(optiStats['Vertical Misalignment Means'])
    optiStats['AngEnd Final SD'] = np.std(optiStats['AngEnd Means'])

    #Creating dictionary of final stats
    finalStats = {'Vertical Misalignment Final Means': [baseStats['Vertical Misalignment Final Mean'], 
                                                        optiStats['Vertical Misalignment Final Mean']],
                  'Vertical Misalignment Final SD': [baseStats['Vertical Misalignment Final SD'], 
                                    optiStats['Vertical Misalignment Final SD']],
                  'Vertical Misalignment T-test': 0,
                  'Vertical Misalignment F-test': 0,
                  'AngEnd Final Means': [baseStats['AngEnd Final Mean'], 
                                         optiStats['AngEnd Final Mean']],
                  'AngEnd Final SD': [baseStats['AngEnd Final SD'], 
                                      optiStats['AngEnd Final SD']],
                  'AngEnd T-test': 0,
                  'AngEnd F-test':0
                 }


    #defining the dataframes to later print as tables
    data1 = pd.DataFrame(data = baseStats)
    data2 = pd.DataFrame(data = optiStats)
    data3 = pd.DataFrame(data = finalStats)
    
    vertMisT = stats.ttest_rel(data1['Vertical Misalignment Means'], data2['Vertical Misalignment Means'])
    data3['Vertical Misalignment T-test'] = vertMisT[1]
    angMisT = stats.ttest_rel(data1['AngEnd Means'], data2['AngEnd Means'])
    data3['AngEnd T-test'] = angMisT[1]
    
    data1.index = ["Trial 1", "Trial 2", "Trial 3", "Trial 4", "Trial 5", "Trial 6", "Trial 7", "Trial 8", "Trial 9", "Trial 10"]
    data2.index = ["Trial 1", "Trial 2", "Trial 3", "Trial 4", "Trial 5", "Trial 6", "Trial 7", "Trial 8", "Trial 9", "Trial 10"]
    data3.index = ["Baseline", "Optivisor"]

    x = np.array(data1['Vertical Misalignment Means'])
    y = np.array(data2['Vertical Misalignment Means'])
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    data3['AngEnd F-test'] = f
    return data1, data2, data3
```
